# Bank-Statement-Analysis

In this Dataset I took my personal bank account data to analyze, It is the year of 1 Sep 2022 to 31 Aug 2023. It having great analysis of the data

## Information About Data:

Transaction Date : This column contain the date of transaction.

Narration : This column contain the information about the transaction Party.

Debit : The amount that I Withdrawal from the account.

Credit : The amount that I Deposit from the account.

Balance : Total balance on account.

```python
# import numpy 
import numpy as np

# import pandas
import pandas as pd

#import matplotlib
import matplotlib.pyplot as plt

#import seaborabsn
import seaborn as sns

# to supress warning
from warnings import filterwarnings
filterwarnings('ignore')

# Display all rows of the dataframe
pd.options.display.max_rows = None
plt.rcParams['figure.figsize'] = [15,8]
```

## Load the data (BOI Bank Statement) 

```python
# load the data(SBI bank statement)
df_BOI = pd.read_excel("bank.xlsx")
```

```python
df_BOI.drop(['Unnamed: 2'], axis=1,inplace = True)
```

## To replace the comma from the deposite column. 

```python
df_BOI['Debit'].unique()
```

```python
df_BOI['Debit'].replace('-','0',inplace = True)
```

```python
df_BOI['Credit'].unique()
```

```python
df_BOI['Credit'].replace('-','0',inplace = True)
```

## Change the data type of following column.

```python
df_BOI["Credit"] = df_BOI['Credit'].astype(float)

df_BOI["Debit"] = df_BOI['Debit'].astype(float)
```

```python
df_BOI.info()
```

## Total income and Total Expense

```python
Total_income = df_BOI['Credit'].sum()
Total_Expense = df_BOI['Debit'].sum()

print("Total_Income :" ,Total_income)

print("Total_Expense :" ,Total_Expense)
```

## Monthly spending patterns

```python
monthly_expenses = df_BOI.groupby(df_BOI['Transaction Date'].dt.month)['Debit'].sum()
monthly_expenses.plot(color = "red",marker='o')
plt.show()
```

## Observation

In the month of May has more spending as compare to other month

## Highest Expenses

```python
top_expenses = df_BOI.nlargest(10, 'Debit')
top_expenses[["Transaction Date","Narration","Debit"]]
```

## Average Monthly Balance:

```python
# Calculate the average monthly balance based on the closing balance for each month
df_BOI['Month'] = df_BOI['Transaction Date'].dt.month

average_monthly_balance = df_BOI.groupby('Month')['Balance'].mean()

average_monthly_balance.plot(kind = "bar",color = "green")

plt.show()
```

## Observation

In the following Four month the bank balance is more than other month.
 

## Income vs Expenses Over Time
 

```python
# Visualize the trend of your income and expenses over the months.

monthly_income = df_BOI.groupby('Month')['Credit'].sum()
monthly_expenses = df_BOI.groupby('Month')['Debit'].sum()

plt.figure(figsize=(10, 6))
plt.plot(monthly_income.index, monthly_income, label='Income', marker='o')
plt.plot(monthly_expenses.index, monthly_expenses, label='Expenses', marker='o')

plt.xlabel('Month')
plt.ylabel('Amount')
plt.title('Income vs. Expenses Over Time')

plt.legend()
plt.show()
```

## Observation

The income and expense rise qucikly between January to March. After this it is slowly down and Quickly increased again in month of May.

The May month expense is more than other month.

## The Week Analysis:

```python
# Check if there's any pattern in your transactions based on the day of the week
#The dt. day_name() function is used to get the day names of the DateTimeIndex with specified locale

df_BOI['Day_of_Week'] = df_BOI['Transaction Date'].dt.day_name()
transactions_by_day = df_BOI.groupby('Day_of_Week').size()

plt.figure(figsize=(8, 6))
transactions_by_day.plot(kind='bar',color = "red")

plt.xlabel('Day of the Week')
plt.ylabel('Number of Transactions')
plt.title('Transactions by Day of the Week')

plt.show()
```

## Observation

The number of transactions are the highest in Wednesday and Saturday. 

## Top Income Transactions

```python
# If the 'Narration' column contains details about income sources, you can identify

top_income_sources = df_BOI.nlargest(5, 'Credit')
top_income_sources[["Transaction Date","Narration","Credit"]]
```

## Net Cash Flow:

```python
# Calculate the net cash flow for each month (income minus expenses).

df_BOI['Net_Cash_Flow'] = df_BOI['Credit'] - df_BOI['Debit']

monthly_cash_flow = df_BOI.groupby('Month')['Net_Cash_Flow'].sum()
monthly_cash_flow.plot(kind = "bar",color = "red")
plt.show()
```

## Observation

In February and October month the deposit amount is more than other.

In March and November month the expenses are higher.

## Transaction Frequency Analysis

```python
# Analyze the frequency of transactions during the year.

transaction_frequency = df_BOI.groupby('Month').size()

plt.figure(figsize=(8, 6))
transaction_frequency.plot(kind='bar',color = "red")
plt.xlabel('Month')
plt.ylabel('Number of Transactions')
plt.title('Transaction Frequency Analysis')
plt.show()
```

## Observation

The number of transactions are higher in the month of February and August.

## Monthly Withdrawal and Deposit Amounts

```python
plt.figure(figsize=(10,6))
withdrawal_per_month = df_BOI.groupby('Month')['Debit'].sum()
deposit_per_month = df_BOI.groupby('Month')['Credit'].sum()

deposit_per_month.plot(kind = "line",color = "green",marker='o')
withdrawal_per_month.plot(kind = "line",color = "red",marker='o')

plt.legend()
plt.show()
```

## Transaction Volume Trend

```python
# Visualize the trend of the number of transaction over the months.

monthly_transaction_count = df_BOI.groupby('Month').size()

plt.figure(figsize=(10,6))
monthly_transaction_count.plot(kind = 'line',marker = 'o',color = 'blue')
plt.xlabel('Month')
plt.ylabel('Number of Transactions')
plt.title('Transaction Volume Trend')
plt.show()
```

## Monthly Withdrawal vs Deposit Comparison


```python
# Compare the monthly total withdrawal and deposit amounts.

monthly_withdrawals = df_BOI.groupby('Month')['Debit'].sum()
monthly_deposits = df_BOI.groupby('Month')['Credit'].sum()

plt.figure(figsize=(10, 6))
plt.bar(monthly_withdrawals.index, monthly_withdrawals, label='Withdrawals')
plt.bar(monthly_deposits.index, monthly_deposits, label='Deposits')

plt.xlabel('Month')
plt.ylabel('Amount')
plt.title('Monthly Withdrawal vs. Deposit')
plt.legend()
plt.show()
```

## Frequency of Expense in a Month

```python
# Analyze your expenses on different days of the month.

expenses_by_day_of_year = df_BOI.groupby(df_BOI['Transaction Date'].dt.day)['Debit'].sum()

plt.figure(figsize=(10, 6))
plt.plot(expenses_by_day_of_year.index, expenses_by_day_of_year, marker='o')
plt.xlabel('Day of the Month')
plt.ylabel('Expenses')
plt.title('Frequency of Expense in a Month')
plt.show()
```
